# Telco Customer Churn — Data Cleaning

## Objective

The goal of this notebook is to clean and prepare the Telco Customer Churn dataset before analysis.

This notebook focuses on:

- loading the dataset
- inspecting its structure
- identifying missing values and data quality issues
- correcting data types
- creating cleaned variables for later analysis

A clean dataset is essential to ensure accurate business, financial, SQL, and predictive analysis.

```python
import pandas as pd
import numpy as np
```

## Load Dataset

We begin by loading the Telco Customer Churn dataset.

```python
df = pd.read_csv("../data/Telco-Customer-Churn.csv")
df.head()
```

## Dataset Shape

We first check the number of rows and columns in the dataset.

```python
print("Dataset shape:", df.shape)
```

## Column Names

We list all columns to understand the dataset structure.

```python
df.columns.tolist()
```

## Data Types and General Structure

We inspect data types and non-null counts to identify possible cleaning needs.

```python
df.info()
```

## Sample of the Dataset

We review a small sample of rows to better understand the values stored in each column.

```python
df.sample(5, random_state=42)
```

## Check for Duplicate Rows

Duplicate rows can distort analysis, so we verify whether any complete duplicates exist.

```python
duplicate_count = df.duplicated().sum()
print("Number of duplicate rows:", duplicate_count)
```

## Remove Duplicate Rows if Needed

If duplicates exist, they should be removed. Otherwise, the dataset remains unchanged.

```python
df = df.drop_duplicates().copy()
print("Dataset shape after duplicate check:", df.shape)
```

## Missing Values

We check for missing values across all columns.

```python
df.isnull().sum()
```

## Check for Blank Strings

In some datasets, missing values are stored as blank spaces instead of true null values.

This is especially important for the `TotalCharges` column in this dataset.

```python
blank_strings = (df == " ").sum()
blank_strings[blank_strings > 0]
```

## Investigate TotalCharges

The `TotalCharges` column should be numeric, but in this dataset some rows contain blank strings.

These rows often correspond to very new customers with little or no billing history.

```python
df[df["TotalCharges"] == " "].head()
```

## Convert TotalCharges to Numeric

We convert `TotalCharges` to numeric format.

Invalid values are turned into missing values using `errors='coerce'`.

```python
df["TotalCharges"] = pd.to_numeric(df["TotalCharges"], errors="coerce")
df["TotalCharges"].head()
```

## Missing Values After Conversion

After converting `TotalCharges`, we check how many missing values were created.

```python
df.isnull().sum()
```

## Remove Rows with Missing TotalCharges

Because only a small number of rows are affected, we remove them to keep the dataset clean and consistent for later analysis.

```python
rows_before = len(df)

df = df.dropna(subset=["TotalCharges"]).copy()

rows_after = len(df)
rows_removed = rows_before - rows_after

print("Rows before cleaning:", rows_before)
print("Rows after cleaning:", rows_after)
print("Rows removed:", rows_removed)
```

## Final Missing Value Check

We confirm that the dataset no longer contains missing values.

```python
df.isnull().sum()
```

## Review Unique Values in Categorical Columns

Before analysis, it is useful to understand the categories present in the dataset.

```python
categorical_cols = df.select_dtypes(include="object").columns.tolist()
categorical_cols
```

```python
for col in categorical_cols:
    print(f"\n{col}")
    print(df[col].unique())
```

## Standardize Churn Variable

To make the dataset easier to use in analysis and modeling, we create a binary churn flag:

- `Yes` → 1
- `No` → 0

```python
df["Churn_Flag"] = df["Churn"].map({"Yes": 1, "No": 0})
df[["Churn", "Churn_Flag"]].head()
```

## Create Tenure Groups

Customer tenure is easier to analyze when grouped into broader lifecycle categories.

```python
df["TenureGroup"] = pd.cut(
    df["tenure"],
    bins=[0, 12, 24, 48, 72],
    labels=["0-12 months", "13-24 months", "25-48 months", "49-72 months"],
    include_lowest=True
)

df[["tenure", "TenureGroup"]].head()
```

## Create Monthly Charges Groups

We group monthly charges into quartiles to simplify pricing analysis later.

```python
df["MonthlyChargesGroup"] = pd.qcut(
    df["MonthlyCharges"],
    q=4,
    labels=["Low", "Medium", "High", "Very High"]
)

df[["MonthlyCharges", "MonthlyChargesGroup"]].head()
```

## Validate New Columns

We verify that the engineered columns were created correctly.

```python
df[["Churn", "Churn_Flag", "tenure", "TenureGroup", "MonthlyCharges", "MonthlyChargesGroup"]].head(10)
```

We confirm the final structure of the cleaned dataset.

To keep the workflow organized, we save the cleaned version of the dataset for use in later notebooks.

## Cleaning Summary

The dataset preparation process included:

- checking dataset size and structure
- verifying duplicates
- identifying blank values in `TotalCharges`
- converting `TotalCharges` to numeric format
- removing rows with missing `TotalCharges`
- creating new analytical variables:
  - `Churn_Flag`
  - `TenureGroup`
  - `MonthlyChargesGroup`

The dataset is now ready for exploratory, business, financial, SQL, and predictive analysis.
